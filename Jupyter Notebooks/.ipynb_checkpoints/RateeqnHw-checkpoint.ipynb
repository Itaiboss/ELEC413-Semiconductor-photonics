```python
import numpy as np
L = 140e-6 #um
W = 2e-6 #um 
T = 211e-9 #nm 
n = 3.5
alpha = 30e2 #cm-1
eta = .5
T_s=3e-9 #ns 
Volt= 1.6 #V 
QW_T = 6e-9 #nm
QW_N =2
q= 1.602176634e-19
h= 6.62607015e-34
c = 3e8
# L = 190e-6
# W = 2e-6
# T = 501e-9
# n = 3.3
# QW_N = 4
# QW_T = 6e-9
# alpha = 30e2
# eta = 0.5
# T_s = 3.0e-9
# Volt = 1.6


#Gain region 
R = ((n - 1) / (n + 1))**2
alpha_m = -1/(L) * np.log(R)
print("Alpha m = ", alpha_m)
alpha_tot = alpha_m + alpha

V_cav= L*W*T
V_active = L * W * QW_N * QW_T
CF = V_active / V_cav
print("Confinement Factor = ", CF)

T_p = ((alpha_tot)*c/n)**-1
print("photon lifetime = ", T_p)

Gth = alpha_tot
#n/c*1/(CF*T_p)

Gmat = Gth / CF

N = 4e24 #cm−3
lam = 1475e-9 #nm 
# N=I*eta*T_s/Q

FSR = lam**2/(2*n*L)
# I_th = (q * N_th * V_gain) / (eta_i * tau_s)

I_th = q*N*V_active/(eta*T_s)


f=c/lam

eta_ext = eta*(alpha_m)/(alpha+alpha_m)
#.02A current 
Pout = eta_ext*h*f/q*(.02-I_th)/2
wpe = Pout/(.02*Volt)

print(f'Gth:{Gth*1e-2}')
print(f'Gmat:{Gmat*1e-2}')
print(f't_photon:{T_p*1e12}')
print(f'FSR:{FSR*1e9}')
print(f'I_th:{I_th*1e3}')
print(f'Pout:{Pout*1e3}')
print(f'Wall plug efficiency:{wpe*100}')
```

```python
#Q4.2
import numpy as np
L = 110e-6 #um
W = 2e-6 #um 
T = 373e-9 #nm 
n = 3.3
alpha = 15e2 #cm-1
eta = .5
T_s=3e-9 #ns 
Volt= 1.5 #V 
QW_T = 6e-9 #nm
QW_N =4
q= 1.602176634e-19
h= 6.62607015e-34
c = 3e8

#Gain region 
R = ((n - 1) / (n + 1))**2
alpha_m = -1/(L) * np.log(R)
print("Alpha m = ", alpha_m)
alpha_tot = alpha_m + alpha

V_cav= L*W*T
V_active = L * W * QW_N * QW_T
CF = V_active / V_cav
print("Confinement Factor = ", CF)

T_p = ((alpha_tot)*c/n)**-1
print("photon lifetime = ", T_p)

Gth = alpha_tot
#n/c*1/(CF*T_p)

Gmat = Gth / CF

N = 4e24 #cm−3
lam = 1460e-9 #nm 
# N=I*eta*T_s/Q

FSR = lam**2/(2*n*L)
# I_th = (q * N_th * V_gain) / (eta_i * tau_s)
B= c/n*Gmat
print(B)
I_th = q*N*V_active/(eta*T_s)

print(f'I_th (mA):{(n/c*1/(CF*B*T_p)+N)/(eta*T_s)*q*V_active*1e3}')

I=5*I_th
P0 = (eta*T_p*CF)/(q*V_active)*(I-I_th)
print(P0)
P0 = (1/B*(eta*I/(q*V_active)-(N/T_s)))
print(P0)

B = 3.9286*1e-20
w = 1/(2*np.pi)*np.sqrt(B*c/n*P0/T_p)
print(f'F:{w*1e-9}')
print(f'Gth:{Gth*1e-2}')
print(f'Gmat:{Gmat*1e-2}')
print(f't_photon:{T_p*1e12}')
print(f'FSR:{FSR*1e9}')
print(f'I_th:{I_th*1e3}')
```

```python
import numpy as np

# Given parameters
W = 2e-6  # Width of the active region in meters
t_active = 30e-9  # Active thickness in meters (5x6nm quantum wells)
L = 100e-6  # Length in meters
t_optical = 1e-6  # Optical mode height in meters
n_g = 3.6  # Group index
c = 3e8  # Speed of light in m/s

# Compute active volume (in m^3)
V_active = W * L * t_active
print(V_active)
# Compute optical volume (in m^3)
V_optical = W * L * t_optical

# Mode confinement factor (Gamma)
Gamma = V_active / V_optical

# Display the mode confinement factor
print(f'Mode Confinement Factor (Γ): {Gamma}')

# Modal gain coefficient (G0) assumed to be known (let's assume a typical value)
G0 = 1e5  # Example modal gain coefficient in 1/s

# Solve for the gain coefficient B (in m^2)
B = G0 * V_optical / (c / n_g)

# Display the gain coefficient B
print(f'Gain Coefficient B: {B:.2e} m^2')
```
