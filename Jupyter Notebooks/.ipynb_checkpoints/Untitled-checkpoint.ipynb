## Interfermoter Model 

From SPD (Silicon Photonics Design) Section 4.3 \
\
Let the input intensity be $I_i$, with electric field $E_i$. At the output of the first Y-branch (the splitter), with top and bottom branch fields:
$$ E_1 = \frac{E_i}{\sqrt{2}} ,E_2 = \frac{E_i}{\sqrt{2}}. $$

The propagation of light in the waveguides is described by:
$$ \beta_1 = \frac{2\pi n_1}{\lambda}, \quad \beta_2 = \frac{2\pi n_2}{\lambda}, $$
and the waveguides have lengths $L_1$ and $L_2 = L_1 + L$, with propagation losses $\alpha_1$ and $\alpha_2$ (for intensity, $\alpha/2$ for electric field). At the end of the two waveguides (at the input to the combiner Y-branch), the fields are:
$$ E_{o1} = E_1 e^{-i\beta_1 L_1 - \frac{\alpha_1}{2} L_1}, $$
$$ E_{o2} = E_2 e^{-i\beta_2 L_2 - \frac{\alpha_2}{2} L_2}. $$

The output of the second Y-branch (the combiner) is:
$$ E_o = \frac{1}{\sqrt{2}} (E_{o1} + E_{o2}) = \frac{E_i}{2} \left(e^{-i\beta_1 L_1 - \frac{\alpha_1}{2} L_1} + e^{-i\beta_2 L_2 - \frac{\alpha_2}{2} L_2}\right). $$

The intensity at the output is thus:
$$ I_o = \frac{I_i}{4} \left| e^{-i\beta_1 L_1 - \frac{\alpha_1}{2} L_1} + e^{-i\beta_2 L_2 - \frac{\alpha_2}{2} L_2} \right|^2. $$

For simplicity, we assume that the total losses are negligible (if losses are to be accounted for, the equation can be evaluated numerically). In this case, after some trigonometry, the equation simplifies to:
$$ I_o = \frac{I_i}{4} \left[ 2 + 2 \cos(\beta_1 L_1 - \beta_2 L_2) \right], $$
$$ I_o = I_i \cos^2 \left( \frac{\beta_1 L_1 - \beta_2 L_2}{2} \right), $$
$$ I_o = \frac{I_i}{2} \left[ 1 + \cos(\beta_1 L_1 - \beta_2 L_2) \right]. $$

The output of the interferometer is thus a sinusoidally varying function of wavelength (via $\beta_1$ and $\beta_2$) for an imbalanced interferometer ($L_1 \neq L_2$). The period is termed the free spectral range, and for identical waveguides, it is determined by:
$$ \text{FSR [Hz]} = \frac{c}{n_g L}, $$
$$ \text{FSR} = \frac{\lambda^2}{n_g L}. $$

```python
import scipy.constants as consts 
import numpy as np
#Desired FSR Gap 25Ghz
FSR = 25*10**9 
c = consts.c
#From Modeling 
#C1
#.35 .345 .34 .335 .330 .325
#ng = np.array([4.493840, 4.506618, 4.520679, 4.537082, 4.555090, 4.573975])

#C2
ng = 4.77116

# Calculate dl
dl = c / (ng * FSR)

print(dl*10**3)
```
